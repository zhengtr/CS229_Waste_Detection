```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab

pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
dataDir = '../data'
batchNum = 0
annotationFile = '{0}/annotations_{1}_train.json'.format(dataDir, batchNum)
```

```python
coco = COCO(annotationFile)
```

```python
categories = coco.loadCats(coco.getCatIds())
categoryNames = [cat['name'] for cat in categories]
categoryIds = coco.getCatIds(categoryNames)
categoryNames
```

```python
imageIds = []
for id in categoryIds:
    imageIds.extend(coco.getImgIds(catIds=id))
```

```python
imageId = np.random.randint(0, len(imageIds))
image = coco.loadImgs(imageIds[imageId])[0]
I = io.imread("{0}/{1}".format(dataDir, image["file_name"]))
plt.axis("off")
plt.imshow(I)
plt.show()
```

```python
plt.imshow(I); plt.axis("off")
annotationIds = coco.getAnnIds(imgIds=image["id"], catIds=categoryIds, iscrowd=None)
annotations = coco.loadAnns(annotationIds)
coco.showAnns(annotations, draw_bbox=True)
```
